```
import requests
from bs4 import BeautifulSoup
import csv
url = 'https://bicicosmetics.vn/collections/hang-moi-ve?page='
num_page=0 
num_data=0
with open('data.csv', 'w') as csv_file:
  writer = csv.writer(csv_file)
  writer.writerow(['TenSP', 'Ma','Danh Muc','Xuat Xu','Thuong Hieu','Giam gia','Gia ban'])
  while ( num_data<1000 ):
      num_page+=1
      url = url+str(num_page)
      r1 = requests.get(url)
      coverpage = r1.content
      soup = BeautifulSoup(coverpage, 'html5lib')
      num_data+=1
      new_feeds = soup.find('div', class_="content-product-list product-list filter clearfix")
      if new_feeds is not None :
        new_feed = new_feeds.find_all('a')  
        #print(new_feed)
        for i in set(new_feed) :
          title = i.get('title')
          link = i.get('href')
          next_link = 'https://bicicosmetics.vn'+link

          r2 = requests.get(next_link)
          coverpage2 = r2.content
          soup2 = BeautifulSoup(coverpage2,'html.parser')
          prices = soup2.find('span', class_="pro-price")
          price = prices.get_text()
          discounts = soup2.find('span',class_="pro-sale")
          if discounts is not None :
            discount = discounts.get_text()
          else : discount = None
          Codes = soup2.find('div',class_="product_meta_wrapper").find('span',class_="sku")
          Code = Codes.get_text()
          index_product = soup2.find('div',class_="product_meta_wrapper").find_all('tr')
          Menus = index_product[1].find_all('td')
          Menu =''
          z = Menus[1].find_all('a')
          for i in range(len(z)):
            Menu = Menu+z[i].get_text()
          Countrys = index_product[2].find_all('td')
          Country = Countrys[1].get_text()
          Brands = index_product[3].find_all('td')
          Brand = Brands[1].get_text()
          print('Title: {} - Code: {} - Menu: {}  -  Country: {} -   Brand: {}  -  discount: {} , price: {}'.format(title ,Code,Menu,Country,Brand,discount, price))

          writer.writerow([title ,Code,Menu,Country,Brand,discount, price])
        else : num_data+=1
 
```
